##  What is the Iris Dataset? <a name="WhatIs"></a>

The Iris flower data set,also known as Fisher's Iris data set is a multivariate data set introduced by the British statistician and biologist Ronald Fisher. 

The data set consists of 50 samples from each of three species of Iris,the setosa, the virginica and the versicolor. Four features were measured from each sample, the length and width of both the sepals and petals. Studying the combination of these four features, Fisher developed a linear discriminant model to distinguish one species from another.

## Getting the Iris Dataset <a name="getting"></a>

Import the Iris Dataset from the online source and assign it an identify to be used again at a later stage in the program.

```python
url =  "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
headings = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'species']

iris = pd.read_csv(url, names=headings)
```

## Visualizaing the Data <a name="visual"></a>

The next step is to graph the data for each variable measured so we can try and see the differences between each type of species

```python
g = sns.violinplot(x='species', y='sepal-length', data=iris, inner='quartile')
plt.show()
g = sns.violinplot(x='species', y='sepal-width', data=iris, inner='quartile')
plt.show()
g = sns.violinplot(x='species', y='petal-length', data=iris, inner='quartile')
plt.show()
g = sns.violinplot(x='species', y='petal-width', data=iris, inner='quartile') 
plt.show()
```

In the above graphs we are able to visualise each variable on each species. We are also able to see how the species compare to eachother across each variable.

We will now take a look at how the species look when we add the variables together. This is shown in the graphs below.

```python
g = sns.pairplot(iris, hue='species', markers='.')
plt.show()
```

## Using Neural Networks to Estimate Species

After seeing the data displayed in the above graphs, it is time we create our own models of the data and estimate their accuracy using neural networks create a validation dataset, create a test harness, create models to predict the correct species type and finally choose the model that is the best at matching the species.


```python
array = iris.values
X = array[:,0:3]
Y = array[:,3]

validation_size = 0.40
seed = 7
train_x, validate_x, train_y, validate_y = model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)

print(train_x.shape)
print(train_y.shape)
print(validate_x.shape)
print(validate_y.shape)
```

We now have the split dataset. The training data will be used to train the models while the validation dataset will be used to validate the accuracy of the models.
